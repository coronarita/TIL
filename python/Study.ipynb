```python
# 1. 버블정렬

def bubble_sort1(data):
    for idx in range(len(data)-1):
        swap = False
        for idx2 in range(len(data)-idx-1):
            if data[idx2] > data[idx2+1]:
                data[idx2], data[idx2+1] = data[idx2+1], data[idx2]
                swap = True
        if swap == False:
            break
    return data
        
```

```python
%%timeit
bubble_sort1(data_list)
```

```python
# 2. 삽입 정렬
def insertion_sort1(data):
    for idx in range(len(data)-1):
        for idx2 in range(idx+1, 0, -1):
            if data[idx2]<data[idx2-1]:
                data[idx2], data[idx2-1] = data[idx2-1], data[idx2]
    return data
```

```python
%%timeit
insertion_sort1(data_list2)
```

```python
# 3. 선택 정렬
def selection_sort1(data):
    for stand in range(len(data)-1):
        lowest = stand
        for idx in range(stand+1, len(data)):
            if data[lowest] > data[idx]:
                lowest = idx
        data[lowest], data[stand] = data[stand], data[lowest]
    return data
```

```python
%%timeit
selection_sort1(data_list3)
```

```python
# 4. 퀵 정렬

def qsort(data):
    if len(data)<=1:
        return data
    
    pivot = data[0]
    
    left = [item for item in data[1:] if pivot > item]
    right = [item for item in data[1:] if pivot < item]
    
    return qsort(left) + [pivot] + qsort(right)
```

```python
# 5. 병합정렬 - split and merge !

def merge_sort(data):
    if len(data) <=1 :
        return data
    mid = len(data) // 2
    left = merge_sort(data[:mid])
    right = merge_sort(data[mid:])
    # print(left, right)
    return merge(left, right)

def merge(left, right):
    # 3개의 경우가 있음
    merged = list()
    lp, rp = 0, 0
    
    while len(left) > lp and len(right) > rp : 
        if left[lp] > right[rp] :
            merged.append(right[rp])
            rp += 1
        else : 
            merged.append(left[lp])
            lp += 1
    
    while len(left) > lp : 
        merged.append(left[lp])
        lp += 1
    
    while len(right) > rp :
        merged.append(right[rp])
        rp += 1
        
    return merged
```

```python
merge_sort(data_list5)
```

```python
# binary search
def binary_search(data, search):
    if len(data) == 1 and search == data[0] : 
        return True
    if len(data) == 1 and search != data[0] : 
        print("Case1")
        return False
    if len(data) == 0 :
        print("Case2")
        return False 
    
    mid = len(data) // 2
    if search == data[mid] :
        return True
    else :
        if search > data[mid] : 
            return binary_search(data[mid:], search)
        else :
            return binary_search(data[:mid], search)
    
```
